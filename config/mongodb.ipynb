```python
df_customers = pd.read_csv(f'{BASE_DIR}/data/mongodb/customers.csv', sep=',')
df_customers.head()
```

```python
df_employees = pd.read_csv(f'{BASE_DIR}/data/mongodb/employees.csv', sep=',')
df_employees.head()
```

```python
df_products = pd.read_csv(f'{BASE_DIR}/data/mongodb/products.csv', sep=',')
df_products.head()
```

```python
from pymongo import MongoClient
client = MongoClient('localhost', 27017)
db_mongo = client['retail_db']
```

```python
customers_data = df_customers.to_dict(orient='records')
employees_data = df_employees.to_dict(orient='records')
products_data = df_products.to_dict(orient='records')
```

```python
db_mongo.customers.insert_many(customers_data)
db_mongo.employees.insert_many(employees_data)
db_mongo.products.insert_many(products_data)
```

```python
customers_col = db_mongo.customers
employees_col = db_mongo.employees
products_col = db_mongo.products
```

```python
jakarta_customers = customers_col.find({'city': 'West Casey'})

for customer in jakarta_customers.limit(5):
    print(customer)
    print()
```

```python
big_spenders = customers_col.find({'total_spent': {'$gt': 1000}})

for customer in big_spenders.limit(5):
    print(customer)
```

```python
pipeline = [
    {
        '$group': {
            '_id': '$membership_level',
            'count': {'$sum': 1}
        }
    }
]

membership_counts = customers_col.aggregate(pipeline)

for item in membership_counts:
    print(item)
```

```python
expensive_products = products_col.find({'price': {'$gt': 400}})

for product in expensive_products.limit(5):
    print(product)
```

```python
high_salary_employees = employees_col.find({'salary': {'$gt': 10000}})

for emp in high_salary_employees.limit(5):
    print(emp)
```

```python
recent_logins = customers_col.find().sort('last_login', -1)

for customer in recent_logins.limit(5):
    print(customer)
```

```python
customer = customers_col.find_one({'email': 'some_email@example.com'})

print(customer)
```

## INDEXING ANALYSIS

```python
from pymongo import MongoClient
from time import perf_counter

client = MongoClient('mongodb://localhost:27017/')
db = client['retail_db']
customers_col = db.customers

# Timing without index
start = perf_counter()
result = customers_col.find({'city': 'West Casey'})
docs = list(result)  # Jangan lupa fetch result
end = perf_counter()

print(f"Query Time Without Index: {end - start:.6f} seconds")

# Indexing
customers_col.create_index('city')

# Timing with index
start = perf_counter()
result = customers_col.find({'city': 'West Casey'})
docs = list(result)
end = perf_counter()

print(f"Query Time With Index: {end - start:.6f} seconds")
```
