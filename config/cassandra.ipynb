```python
import uuid
from datetime import datetime

def uuid_parse(x):
    if pd.isnull(x):
        return None
    return uuid.UUID(x) if not isinstance(x, uuid.UUID) else x

def date_parse(x):
    if pd.isnull(x):
        return None
    if isinstance(x, str):
        try:
            return datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f')
        except ValueError:
            return datetime.strptime(x, '%Y-%m-%d %H:%M:%S')
    return x

def int_parse(x):
    if pd.isnull(x):
        return None
    return int(x)

def float_parse(x):
    if pd.isnull(x):
        return None
    return float(x)

def str_parse(x):
    if pd.isnull(x):
        return None
    return str(x)
```

```python
from cassandra.query import BatchStatement, ConsistencyLevel
from tqdm import tqdm

def insert_dataframe_to_cassandra_batch(df, table_name, columns, parse_funcs, batch_size=50):
    query = f"INSERT INTO {table_name} ({', '.join(columns)}) VALUES ({', '.join(['?' for _ in columns])})"
    prepared = session.prepare(query)

    batch = BatchStatement(consistency_level=ConsistencyLevel.QUORUM)
    count = 0

    for idx, row in tqdm(df.iterrows(), total=len(df), desc=f'Inserting into {table_name} with batch'):
        try:
            values = tuple(parse(row[col]) for col, parse in zip(columns, parse_funcs))
            batch.add(prepared, values)
            count += 1
            
            if count % batch_size == 0:
                session.execute(batch)
                batch.clear()
        
        except Exception as e:
            print(f"Error inserting row {row.to_dict()} with error: {e}")
    
    # Insert remaining batch
    if batch:
        session.execute(batch)

# Insert from dataframes
insert_dataframe_to_cassandra_batch(df_transactions, 'sales_transactions',
    ['transaction_id', 'user_id', 'product_id', 'purchase_date', 'quantity', 'total_amount', 'payment_method'],
    [uuid_parse, uuid_parse, uuid_parse, date_parse, int_parse, float_parse, str_parse],
    batch_size=50
)

insert_dataframe_to_cassandra_batch(df_payment_histories, 'payment_histories',
    ['payment_id', 'user_id', 'payment_date', 'amount', 'payment_method'],
    [uuid_parse, uuid_parse, date_parse, float_parse, str_parse],
    batch_size=50
)

insert_dataframe_to_cassandra_batch(df_inventory_logs, 'inventory_logs',
    ['product_id', 'timestamp', 'stock_change', 'description'],
    [uuid_parse, date_parse, int_parse, str_parse],
    batch_size=50
)
```

```python
from cassandra.cluster import Cluster
import time

# Setup Cassandra connection
cluster = Cluster(['127.0.0.1'])
session = cluster.connect('retail_transactions')

table_name = 'payment_histories'
field = 'payment_method'

# 1. Benchmark without index
query = f"SELECT * FROM {table_name} WHERE {field} = 'some_value' ALLOW FILTERING"

start_time = time.time()
rows = session.execute(query)
_ = list(rows)
end_time = time.time()
cassandra_time_no_index = end_time - start_time
print(f"Cassandra - Time WITHOUT Index: {cassandra_time_no_index:.4f} seconds")

# 2. Create Index
session.execute(f"CREATE INDEX IF NOT EXISTS ON {table_name} ({field});")

# 3. Benchmark with index
start_time = time.time()
rows = session.execute(query)
_ = list(rows)
end_time = time.time()
cassandra_time_with_index = end_time - start_time
print(f"Cassandra - Time WITH Index: {cassandra_time_with_index:.4f} seconds")
```
